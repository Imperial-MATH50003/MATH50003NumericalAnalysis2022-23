```julia
using Plots, Test
```

```julia
f = x -> 10x^2 + 3x + 1 + 0.1randn() # quadratic plus normal noise

m = 200
x = range(0,1; length=m) # 200 evenly spaced points between 0 and 1
 = f.(x)


V = [x[k]^j for k=1:m, j=0:2] # x .^ (0:2)'

c = V \  # least squares solve
p = x -> c[1] + c[2]x + c[3]x^2

scatter(x, )
plot!(x, p.(x))
```

```julia
Q,R = qr(V)
# R is reduced R
# Q is a type representing both full and reduced Q
# it acts on both ^m and ^n
Q * [1,2,3]  Q * [1; 2; 3; zeros(m-3)]
```

```julia
Q' # it behaves like the full Q, which is U(m)
R \ (Q')[1:3]  c
```

```julia
f = x -> 10x^2 + 3x + 1 + 0.1randn() # quadratic plus normal noise

m = 200
x = range(0,1; length=m) # 200 evenly spaced points between 0 and 1
 = f.(x)

g = range(0,1; length=1000) # plotting grid

n = 180

V = [x[k]^j for k=1:m, j=0:n-1] # x .^ (0:2)'

c = V \  # least squares solve

V_g = [g[k]^j for k=1:1000, j=0:n-1] # x .^ (0:2)'


scatter(x, )
plot!(g, V_g * c)
```
