```julia
𝐱 = [-1,-0.5,0.1,0.6,1.0]
ℓ =  (k,x) ->  prod((x - 𝐱[j])/(𝐱[k]-𝐱[j]) for j = [1:k-1; k+1:length(𝐱)])

g = range(-1, 1; length=1000)
scatter(𝐱, zero(𝐱); legend=false)
plot!(g, ℓ.(1,g))
plot!(g, ℓ.(2,g))
plot!(g, ℓ.(3,g))
```

```julia
p = x -> dot(exp.(𝐱), [ℓ(k,x) for k=1:length(𝐱)])
g = range(-1, 1; length=1000)
scatter(𝐱, exp.(𝐱); legend=false)
plot!(g, p.(g))
```
