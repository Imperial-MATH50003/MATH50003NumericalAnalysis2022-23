```julia
using ClassicalOrthogonalPolynomials, Plots
```

# Classical OPs: Chebyshev and Legendre

```julia
x = range(-1., 1.; length=1000)
plot(x, acos.(x))
```

```julia
x = range(-1., 1.; length=2000)
T = (n,x) -> cos(n*acos(x))
plot(x, T.(5,x); legend=false)
plot!(x, T.(100,x))
```

```julia
x = range(-1., 1.; length=1000)

n = 10
plot(x, chebyshevt.(n,x); label="Chebyshev") # uses 3-term recurrence
plot!(x, legendrep.(n,x); label="Legendre")
```

# Lagrange polynomials

```julia
𝐱 = [-1,-0.5,0.1,0.6,1.0]
ℓ =  (k,x) ->  prod((x - 𝐱[j])/(𝐱[k]-𝐱[j]) for j = [1:k-1; k+1:length(𝐱)])

g = range(-1, 1; length=1000)
scatter(𝐱, zero(𝐱); legend=false)
plot!(g, ℓ.(1,g))
plot!(g, ℓ.(2,g))
plot!(g, ℓ.(3,g))
```

```julia
p = x -> dot(cos.(2𝐱), [ℓ(k,x) for k=1:length(𝐱)])
g = range(-1, 1; length=1000)
scatter(𝐱, cos.(2𝐱); legend=false)
plot!(g, p.(g))
```
