```julia
x = range(-1.1, 1.1; length=100)
plot(x, chebyshevt.(10,x); label="Chebyshev")
plot!(x, legendrep.(10,x); label="Legendre")
```

```julia
x = range(-1, 1; length=10000)
n = 20
plot(x, chebyshevt.(n,x); label="Chebyshev")
plot!(x, legendrep.(n,x); label="Legendre")
```

```julia
 = [-1,-0.5,0.1,0.6,1.0]
 =  (k,x) ->  prod((x - [j])/([k]-[j]) for j = [1:k-1; k+1:length()])

g = range(-1, 1; length=1000)
scatter(, zero(); legend=false)
plot!(g, .(1,g))
plot!(g, .(2,g))
plot!(g, .(3,g))
```

```julia
p = x -> dot(exp.(), [(k,x) for k=1:length()])
g = range(-1, 1; length=1000)
scatter(, exp.(); legend=false)
plot!(g, p.(g))
```
